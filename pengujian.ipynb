### Fungsi Fungsi

```python
def apply_fft(x, fs, num_samples):
    f = np.linspace(0.0, (fs/2.0), num_samples//2)
    freq_values = np.fft.fft(x)
    freq_values = 2.0/num_samples * np.abs(freq_values[0:num_samples//2])
    return f, freq_values
```

```python
import pywt
import numpy as np
from scipy.signal import butter, lfilter
```

```python
def lowpass_2(x):
  # Cut-off frequency of the filter (in Hz)
  lowcut = 1000

  # Sampling rate of the data (in Hz)
  fs = 50000

  # Filter order
  order = 2

  # Design the band-pass Butterworth filter
  nyquist = 0.5 * fs
  low = lowcut / nyquist
  b, a = butter(order, low, btype='low', analog=False)

  # Apply the filter to the data
  y = lfilter(b, a, x)

  return y
```

```python
def lowpass(x):
  # Cut-off frequency of the filter (in Hz)
  lowcut = 100

  # Sampling rate of the data (in Hz)
  fs = 48000

  # Filter order
  order = 1

  # Design the band-pass Butterworth filter
  nyquist = 0.5 * fs
  low = lowcut / nyquist
  b, a = butter(order, low, btype='low', analog=False)

  # Apply the filter to the data
  y = lfilter(b, a, x)

  return y
```

```python
def highpass(x):
  # Cut-off frequency of the filter (in Hz)
  highcut = 2

  # Sampling rate of the data (in Hz)
  fs = 200

  # Filter order
  order = 1

  # Design the band-pass Butterworth filter
  nyquist = 0.5 * fs
  high = highcut / nyquist
  b, a = butter(order, high, btype='high', analog=False)

  # Apply the filter to the data
  y = lfilter(b, a, x)

  return y
```

#### Preprocessing - Wavelet Transform

```python
import pywt
```

```python
def wavelet_new(sinyal,fs):
  temp = pywt.WaveletPacket(data=sinyal, wavelet='db4', mode='symmetric')
  x = [node.path for node in temp.get_level(3, 'natural')]
  temp_feature = [np.sum((temp[i].data)**2) for i in x]

  for i in x:
    new_wp = pywt.WaveletPacket(data = None, wavelet = 'db4', mode='symmetric',maxlevel=3)
    new_wp[i] = temp[i].data
    reconstructed_signal = new_wp.reconstruct(update = False) # Signal reconstruction from wavelet packet coefficients
    f, c = apply_fft(reconstructed_signal, fs, len(reconstructed_signal))

    z = abs(c)

    # Find  m  highest amplitudes of the spectrum and their corresponding frequencies:
    maximal_idx = np.argpartition(z, -1)[-1:]
    high_amp = z[maximal_idx]
    high_freq = f[maximal_idx]
    feature = high_amp*high_freq
    temp_feature.append(list(high_amp)[0])
    temp_feature.append(list(high_freq)[0])
    temp_feature.append(list(feature)[0])

  return temp_feature
```

```python
def compute_shannon_entropy(signal):
    return -np.nansum(signal**2 * np.log(signal**2))
```

```python
def time_feature(temp,length,col):
  average = [np.mean(temp.iloc[xx:xx+length,col].values)
                if (xx+length<temp.shape[0]) else np.mean(temp.iloc[xx:temp.shape[0],col].values)
                for xx in range(0,temp.shape[0],length)]

  rms = [np.sqrt(np.mean(temp.iloc[xx:xx+length,col].values**2))
                if (xx+length<temp.shape[0]) else np.sqrt(np.mean(temp.iloc[xx:temp.shape[0],col].values**2))
                for xx in range(0,temp.shape[0],length)]

  max_abs = [max(abs(temp.iloc[xx:xx+length,col].values))
                if (xx+length<temp.shape[0]) else max(abs(temp.iloc[xx:temp.shape[0],
                                                                    col].values))
                for xx in range(0,temp.shape[0],length)]

  vpp = [abs(max(temp.iloc[xx:xx+length,col].values))-abs(min(temp.iloc[xx:xx+length,col].values))
                if (xx+length<temp.shape[0]) else 
             abs(max(temp.iloc[xx:temp.shape[0],col].values))-abs(min(temp.iloc[xx:temp.shape[0],col].values))
                for xx in range(0,temp.shape[0],length)]

  shannon = [compute_shannon_entropy(temp.iloc[xx:xx+length,col].values)
                if (xx+length<temp.shape[0]) else 
             compute_shannon_entropy(temp.iloc[xx:temp.shape[0],col].values)
                for xx in range(0,temp.shape[0],length)]

  merged_dat=pd.DataFrame({
      'mean':average,
      'rms':rms,
      'max_abs':max_abs,
      'vpp':vpp,
      'shannon':shannon
  })
  return merged_dat
```

```python
def time_feat(data):
  average=np.mean(data)
  rms=np.sqrt(np.mean(data**2))
  max_abs=max(abs(data))
  vpp=abs(max(data))-abs(min(data))
  shannon=compute_shannon_entropy(data)

  return [average,rms,max_abs,vpp,shannon]
```

```python
def wav_feature_new(dat_1,length,col,fs,label):
  merged_dat = pd.DataFrame()

  for i in range(1,3):
    temp = pd.DataFrame(dat_1.iloc[:,i].values)
    print(temp.shape)
    time_feat = time_feature(temp,length,0)
    # print('hmhmmh')
    print(len(temp.iloc[0:0+length,0].values))
    wavelett = [wavelet_new(temp.iloc[xx:xx+length,0].values,fs)
                if (xx+length<temp.shape[0]) else wavelet_new(temp.iloc[xx:temp.shape[0],0].values,fs)
                for xx in range(0,temp.shape[0],length)]

    # print("bang")
    # fault = [temp.iloc[0,-1] for k in range(0,temp.shape[0],length)]
      
    res_dat = pd.concat([pd.DataFrame(time_feat),pd.DataFrame(wavelett)],axis=1)
    
    merged_dat = pd.concat([merged_dat,res_dat],axis=0)

  merged_dat['fault']=label
  print(merged_dat.shape)
  return merged_dat
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch import sigmoid,tanh
import torch.optim as optim

torch.manual_seed(0)
torch.set_default_dtype(torch.float32)
device = torch.device("cpu")
```

```python
class JANet(nn.Module):
    def __init__(self, inputs, cells, num_outputs, num_timesteps, output_activation=None):
        super(JANet, self).__init__()
        
        self.inputs = inputs
        self.cells = cells
        self.classes = num_outputs
        self.num_timesteps = num_timesteps
        self.output_activation = output_activation
        
        kernel_data = torch.zeros(inputs, 2 * cells, dtype=torch.get_default_dtype())
        kernel_data = torch.nn.init.xavier_uniform_(kernel_data)
        self.kernel = nn.Parameter(kernel_data)
        
        recurrent_kernel_data = torch.zeros(cells, 2 * cells, dtype=torch.get_default_dtype())
        recurrent_kernel_data = torch.nn.init.xavier_uniform_(recurrent_kernel_data)
        self.recurrent_kernel = nn.Parameter(recurrent_kernel_data)
        
        recurrent_bias = np.zeros(2 * cells)
        # chrono initializer
        recurrent_bias[:cells] = np.log(np.random.uniform(1., self.num_timesteps - 1, size=cells))
        recurrent_bias = recurrent_bias.astype('float32')
        self.recurrent_bias = nn.Parameter(torch.from_numpy(recurrent_bias))
        
        self.output_dense = nn.Linear(cells, num_outputs)
        
    def forward(self, inputs):
        h_state = torch.zeros(inputs.size(0), self.cells, dtype=torch.get_default_dtype()).to(device)
        c_state = torch.zeros(inputs.size(0), self.cells, dtype=torch.get_default_dtype()).to(device)
        
        num_timesteps = inputs.size(1)
        
        for t in range(num_timesteps):
            ip = inputs[:, t, :]
            
            z = torch.mm(ip, self.kernel)
            z += torch.mm(h_state, self.recurrent_kernel) + self.recurrent_bias
            
            z0 = z[:, :self.cells]
            z1 = z[:, self.cells: self.cells * 2]
            
            f = sigmoid(z0)
            c = f * c_state + (1. - f) * tanh(z1)
            
            h = c
            
            h_state = h
            c_state = c
        
        preds = self.output_dense(h_state)
        
        if self.output_activation is not None:
            preds = self.output_activation(preds)
        
        return preds
```

```python
def multi_acc(y_pred, y_test):
    y_pred_softmax = torch.log_softmax(y_pred, dim = 1)
    _, y_pred_tags = torch.max(y_pred_softmax, dim = 1)    
    
    correct_pred = (y_pred_tags == y_test).float()
    acc = correct_pred.sum() / len(correct_pred)
    
    acc = torch.round(acc * 100)
    
    return acc
```

```python
def wavelet_new(sinyal,fs):
  temp = pywt.WaveletPacket(data=sinyal, wavelet='db4', mode='symmetric')
  x = [node.path for node in temp.get_level(3, 'natural')]
  temp_feature = [np.sum((temp[i].data)**2) for i in x]

  for i in x:
    new_wp = pywt.WaveletPacket(data = None, wavelet = 'db4', mode='symmetric',maxlevel=3)
    new_wp[i] = temp[i].data
    reconstructed_signal = new_wp.reconstruct(update = False) # Signal reconstruction from wavelet packet coefficients
    f, c = apply_fft(reconstructed_signal, fs, len(reconstructed_signal))

    z = abs(c)

    # Find  m  highest amplitudes of the spectrum and their corresponding frequencies:
    maximal_idx = np.argpartition(z, -1)[-1:]
    high_amp = z[maximal_idx]
    high_freq = f[maximal_idx]
    feature = high_amp*high_freq
    temp_feature.append(list(high_amp)[0])
    temp_feature.append(list(high_freq)[0])
    temp_feature.append(list(feature)[0])

  return temp_feature
```
